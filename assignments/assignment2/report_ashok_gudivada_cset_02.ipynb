## Gaussian Smoothing as a Noise/Scale Tool
---

1. Which σ best reduces grain while keeping facial/scene edges reasonably sharp?

   The optimal σ for reducing grain while maintaining reasonably sharp edges can vary depending on the specific image and the level of noise present:
   - A **smaller σ** (e.g., σ = 1–2) effectively reduces grain while preserving edges.
   - A **larger σ** (e.g., σ > 3) leads to excessive blurring and loss of detail.

3. Compare a Gaussian blur to a same-size box filter. Comment on artifacts (e.g., ringing is not expected
here, but edge fattening and texture smearing differ).

  Both filters smooth the image, but they differ in how they handle edges and textures:
 - **Edge fattening** — the box filter produces more pronounced edge fattening due to its uniform
   weighting, whereas the Gaussian filter's bell-shaped kernel tapers weights toward the edges,
   resulting in softer, more natural-looking transitions.

 - **Texture smearing** — the box filter smears textures more aggressively and introduces blocky
   artifacts, since all pixels within the kernel are weighted equally regardless of distance. The
   Gaussian filter smears textures more gradually, better preserving local structure.

 - **Ringing** — neither filter introduces ringing artifacts in the spatial domain, as both kernels
   are non-negative and contain no oscillatory components.

1. Examine a few different values of k and discuss what you think is the best value to use.

   - **Small k** (e.g., k = 1) — captures fine details but is highly sensitive to noise, producing
     many false edges from grain and scratches.
   - **Medium k** (e.g., k = 3) — balances edge localization and noise suppression; widely used as
     the default Sobel kernel size.
   - **Large k** (e.g., k = 5 or 7) — further smooths responses but may broaden edges and miss
     fine structures.

   **k = 3** is generally the best choice, as it provides a good trade-off between sensitivity to
   true edges and robustness to noise.

<br/>

3. Is there any advantage to a Scharr filter over Sobel?

   Yes — the **Scharr filter** uses optimized kernel weights (e.g., [3, 10, 3] / [−3, 0, 3]) that
   provide more **isotropic gradient estimation** than the standard Sobel kernel. This means the
   Scharr filter is less sensitive to the orientation of edges and produces more accurate gradient
   magnitudes for diagonal edges, making it preferable when rotational accuracy matters.

## Median Filtering for Dust/Speckles

### Median vs Gaussian on Impulse-Like Artifacts

Using kernel size **k = 5**, the median filter and Gaussian smoother were applied to the degraded
image and compared:

![Median Filter](figs/median_old_photo_01_k3.png)
![Gaussian Filter](figs/gaussian_old_photo_01_sigma3.png)

1. **Which filter is better at removing speckles?**

   The **median filter** is significantly better at removing impulse-like artifacts (dust, speckles).
   Because it replaces each pixel with the *median* of its neighborhood, isolated outlier pixels
   (black/white dust specks) are simply discarded rather than averaged in. The Gaussian filter, by
   contrast, blends the outlier into its surroundings, leaving a faint smear where each speck was.

2. **Which is better at preserving edges?**

   The **median filter** also preserves edges better. Since it is a rank-order (non-linear) filter,
   a sharp intensity step remains sharp as long as more than half the pixels in the window lie on
   the same side of the edge. The Gaussian filter is linear and always blurs across boundaries,
   causing measurable edge fattening proportional to σ.

   - Fine facial features and scene textures remain recognisable, unlike k ≥ 7 where
     structural details begin to merge.

# Design your Pipeline 
Design a pipeline (at least 4 steps) using tools from this lab. Example steps include:
1. Contrast normalization (optional)
2. Gaussian smoothing (grain)
3. Median filtering (dust)
4. Scratch mask + morphology
5. Inpainting using cv.inpaint (optional)
6. Final contrast enhancement (hist eq or CLAHE)
Provide a one-paragraph justification of the ordering and parameters. Feel free to research other methods
and add or substitute them in your pipeline.



The pipeline is ordered from global intensity correction to localized defect reconstruction so that each operation works on a progressively cleaner signal. Contrast normalization is applied first to expand faded pixel values across the full dynamic range. This ensures that later gradient calculations and threshold-based detectors operate on consistent intensity distributions rather than compressed tonal regions.

Gaussian smoothing follows to reduce high-frequency film grain while retaining primary structural edges. Lowering this stochastic noise before further processing improves the stability of subsequent filters. The median filter is then applied to remove impulse noise such as dust and salt-and-pepper artifacts. Because the image has already been smoothed, a moderate kernel size is sufficient, which limits edge distortion.

Scratch detection is performed only after global denoising. The multi-scale brightness analysis, Hessian ridge detection, and Difference-of-Gaussians line detection operate on a signal where random grain has been suppressed, reducing false positives. Morphological refinement then consolidates detected responses into a coherent binary mask.

Inpainting is applied once the defect mask is finalized, ensuring that reconstruction targets only confirmed damaged regions. The surrounding pixels used for interpolation have already been normalized and denoised, improving reconstruction consistency. A final CLAHE pass compensates for minor contrast reduction introduced by filtering, restoring local contrast without substantially re-amplifying noise.
